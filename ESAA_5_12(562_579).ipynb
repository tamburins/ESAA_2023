### 추천 시스템의 개요

유튜브, 애플뮤직 등에서 추천시스템은 몇개의 콘텐츠만 클릭해도 그와 유사하거나 연관된 콘텐츠를 줄줄이 추천해주며, 사용자가 관심을 가질만한 콘텐츠는 물론이고 그동안 잊고 지내왔거나 몰랐던 새로운 취향의 콘텐츠까지 소개해준다.

하나의 콘텐츠를 선택했을 때 선택된 콘텐츠와 연관된 추천 콘텐츠가 얼마나 사용자으 ㅣ관심을 끌고 개인에게 맞춘 콘텐츠를 추천했는지는 그 사이트의 평판을 결정하는 중요한 요소로 , 사용자 자신도 몰랐던 취향을 시스템이 발견하고 그에 맞는 콘텐츠를 추천해주는 것이다. 이는 사용자와 추천사이트간 신뢰를 쌓으면서 데이터가 더 많이 축적되고, 추천이 다시 정확해지고 다양한 결과를 얻게되는 좋은 선순환 시스템 구축에 용이하다.

## 03 최근접 이웃 협업 필터링
신작 영화 개봉시 영화를 보러갈지말지 여부는 영화를 보러가는데 소모되는 비용을 감안하여 주변의 지인들에게 후기를 물어 결정하곤 한다.
이와 유사하게 사용자가 아이템에 매긴 평점정보나 상품 구매이력과 같은 사용자 행동양식만을 기반으로 추천을 수행하는 것을 협업 필터링방식ㅇ리ㅏ 한다.

이의 주요 목표는 사용자-아이템 평점 매트릭스와 같은 축적된 사용자 행동 데이터를 기반으로 사용자가 아직 평가하지 않은 아이템을 예측평가하는 것이다. 즉 사용자가 평가하지 않은 아이템에 대해 이미 평가한 다른 아이템을 기반으로 예측 평가를 도출하는 방식이다.

협업 필터링 기반의 추천시스템은 최근접 이웃방식과 잠재 요인 방식으로 나뉘며, 두 방식 모두 사용ㅈ자-아이템 평점 행렬 데이터에만 의지해 추천을 수행한다. 협업 필터링 알고리즘에 사용되는 사용자-아이템 평점 행렬에서 행은 개별사용자, 열은 개별 아이템으로 구성되며 사용자 아이디 행, 아이템 아이디 열 위치에 해당하는 값이 평점을 나타내는 형태가 되야한다.
일반적으로 이러한 사용자-아이템 평점 행렬은 많은 아이템을 열로 가지는 다차원 행렬이며, 사용자가 아이템에 대한 평점을 매기는 경우가 많지 않기 때문에 희소행렬 특성을 갖는다.

최근접 이웃 협업 필터링은 메모리 협업 필터링이라고도 하며, 일반적으로 사용자기반과 아이템 기반으로 다시 나뉠 수 있다.
사용자기반은 당신과 비슷한 고객들이 다음 상품도 구매한 것, 아이템 기반은 이상품을 선택한 고객들은 다음 상품도 구매했다 이다.

사용자 기반 최근접 이웃 방식은 특정 사용자와 유사한 다른 사용자를 Top-n으로 선정해 이 top-n 사용자가 좋아하는 아이템을 추천하는 것으로 특정 사용자와 타 사용자간의 유사도를 측정한 뒤 가장 유사도가 높은 top-n 사용자를 추출해 선호하는 아이템을 추천하는 것이다.

예를 들어 사용자 a는 주요 영화의 평점정보가 사용자 c보다 b와 비슷하므로 사용자a와 b는 상호간 유사도가 매우 높다 할 수 있다. 만약 a에게 아직 보지 못한 두개의 영화인 프로메테우스와 스타워즈 라스트 제다이 중 하나를 추천한다면 c가 재밌게 본 스타워즈 라스트 제다이보다는 a와 유사도가 높은 b가 본 프로메테우스를 추천하는 것이 사용자 기반 최근접 이웃 협업필터링이다.

아이템 기반 최근접 이웃 방식은 그 명칭이 주는 이미지때문에 아이템간 속성을 비교한다 착각할 수 있으나 아이템의 속성과는 상관없이 사용자들이 아이템을 좋아하는지/싫어하는지 평가 척도가 유사한 아이템을 추천하는 기준이 되는 알고리즘이다.
이는 위의 사용자 기반 최근접 이웃데이터세트와 행고 ㅏ열이 서로 반대이다.

아이템 다큰아ㅣ트는 스타워즈 라스트제다이보다 프로메테우스와 사용자들의 평점분포가 훨씬 더 비슷하므로 다크나이트와 프로메테우스는 상호간 아이템유사도가 상대적으로 매우 높다. 따라서 다크나이트를 좋아하는 사용자 d에게 아이템기반 협업 필터링은 d가 아직 관람하지 못한 프로메테우스오 ㅏ스타워즈 라스트제다이 중 프로메테우스를 추천한다.
일반적으로 사용자기반보다는 아이템 기반 협업 필터링이 정확도가 더 높다. 이유는 비슷한 영화를 좋아한다고 해서 사람들의 취향ㅇ ㅣ비슷하다고 판단하기는 어려운 경우가 많다. 유명영화는 취향과 관계없이 많이 보는 경우가 많고, 사용자들이 평점을 매긴 영화의 개수가 많지 않은데다, 이를 기반으로 다른 사람들과의 유사도를 비교하기 어려운 부분도 있다. 따라서 최근접 이웃협업 필터링은 대부분 아이템 기반의 알고리즘을 적용한다.

앞장의 텍스트분석에서 소개된 유사도 측정 방법인 코사인 유사도는 추천시스템의 유사도 측정에 가장 많이 적용된다. ( 희소행렬이기 때문 )

## 04 잠재 요인 협업 필터링

### 잠재 요인 협업 필터링의 이해
잠재요인 협업 필터링은 사용자 아이템 평점 매트릭스 속에 숨어있는 잠재요인을 추출해 추천 예측을 할 수 있게 하는 기법이다. 대규모 다차원 행렬을 svd와 같은 차원 감소 기법으로 분해하는 과정에서 잠재 요인을 추출하는데, 이러한 기법을 행렬분해라 한다.

잠재요인 협업 필터링은 사용자-아이템 평점 행렬 데이터만을 이용해 잠재요인을 끄집어내는 것으로 잠재요인은 명확히 정의할 수 없으나 이를 기반으로 다차원 희소 행렬인 사용자-아이템 행렬 데이터를 저차원 밀집 행렬의 사용자-잠재요인 행렬과 아이템-잠재 요인 행렬의 전치행렬로 분해할 수 있으며, 이렇게 분해된 두 행렬의 내적을 통해 새 예측사용자-아이템 평점 행렬데이터를 만들어 사용자가 아직 평점을 부여하지 않은 아이템에 대해 예측 평점을 생성하는 것이 잠재요인 협력 필터링 알고리즘의 요점이다.

image.png

행렬분해에 의해 추출되는 잠재요인은 예를 들어 영화평점 기반의 사용자-아이템 평점 행렬 데이터라면 영화가 갖는 장르별 특성 선호도라 할 수 있다. 즉 사용자-잠재요인 행렬은 사용자의 영화장르에 대한 선호도로, 아이템 잠재요인 행렬은 영화으 ㅣ장르별 특성값이라 할 수 있다.

다음 그림의 사용자-아이템 평점 행렬에서 사용자의 아이템에 대한 평점을 R(u,i)라 하고, 사용자-잠재요인 행렬을 사용자의 영화장르별 선호도 행렬 P, factor1을 액션 선호도, factor2를 로맨스 선호도라 가정한다.(편의상 영화장르를 이진법으로 나누었다 가정하자)
아이템 잠재요인 행렬은 영화별 여러 장르요소로 구성된 영화의 장르별 요소행렬 Q로 가정하고 factor1은 영화의 action요소값, 팩터2는 로맨스요소값이다. Q(i,k)에서 i는 아이템 아이디,k는 잠재요인 칼럼인 장르별 요소이다. Q는 p와의 내적 계산을 통해 예측평점을 계산하기 ㅜ이해 q의 행과 열 위치를 교환한 q전치로 변환한다.

평점이란 사용자의 특정영화장르에 대한 선호도와 개별 영화의 장르적 특성값을 반영해 결정된다 할 수 있다. 예를 들어 사용자가 액션 영화를 좋아하고 특정 영화가 액션 영화의특성이 매우 크다면 사용자가 해당 영화에 매우 높은 평점을 줄 것이다. 즉 평점은 사용자의 장르별 선호도 벡터와 영화으 ㅣ장르별 특성 벡터를 서로 곱해 만들 수 있다.

마찬가지 방법으로 아직 user1이 평점을 매기지 못한 item2에 대해 예측 평점을 수행해본다.

이처럼 잠재요인 협업 필터링은 숨겨진 잠재요인을 기반으로 분해된 메트릭스를 이ㅛㅇ해 사용자가 아직 평가하지 않은 아이템에 대해 예측 평가를 수행하는 것이다. 사용자-아이템 평점 행렬과 같이 다차원의 메트릭스를 저차원의 메트릭스로 분해하는 기법을 행렬분해라 한다.

### 행렬 분해의 이해
행렬 분해는 다차원 메트릭스를 저차원으로 분해하는 기법으로 SVD, NMF 등이 있다.

분해는 우리말로 인수분해를 의미하는 것으로 일반적으로 하나의 다항식을 두개이상의 좀 더 단순한 인순의 곱을 분해하는 것을 의미한다. 

M x N차원의  m행 n열 행렬은 행렬분해를 통해 사용자 k차원 잠재요인 행렬과 k차원 잠재요인-아이템 행렬 Q.T로 분해할 수 ㅣㅇㅆ다.

즉 R = P*Q.T로 M은 총 사용자수 n은 총 아이템 수, k은 잠재요인의 수, r은 MXN차원의 사용자-아이템 평점행렬ㄹ, p는 사용자와 잠재요인과의 관계값을 가지는 mxk차원의 사용자-잠재요인 행렬, q는 아이템과 잠재요인과의 관계값을 가지는 nXK차원의 아이템-잠재요인 행렬, q.T는 매트릭스의 행과 열값을 교환한 전치행렬이다.

예를 들어 행렬 내에 널값을 많이 갖는 고차원의 희소행렬인 r행렬은 다음 그림과 같이 밀집행렬 p와 q행렬로 분해될 수 있다.

r행렬의 u행 사용자와 i열 아이템 위치에 있는 평점데이터를 r(u,i) 라하면
r(u,i) = Pu * Qi.t
로 유추할 수 있다. Pu는 여기서 u행 사용자의 벡터, Qi.t는 Q행렬의 i행 아이템 벡터의 전치벡터이다.

사용자가 평가하지 않은 아이템에 대한 평점도 잠재요인으로 분해된 p, q행렬을 통해 예측할 수 있다. 

사용자 -아이템 평점행렬의 미정값을 포함한 모든 평점값은 행렬분해를 통해 얻어진 p행렬과 q.t 행렬의 내적을 통해 예측 평점으로 다시 계산할 수 있다.

그렇다면 r을 어떻게 p와 q로 분해할 것인가? 널값이 존재하는한 svd는 적용할 수없으므로 확률적 경사하강법이나 als 방식을 이용하여 svd를 수행한다.

### 확률적 경사하강법을 이용한 행렬 분해
p,q행렬로 계싼된 R 예측행렬값이 실제 행렬r값과 가장 최소으 ㅣ오류를 가질 수 있도록 반복적인 비용함수  최적화를 통해 p,q를 유추하는 것이다.

1. P,Q값을 임의의 값을 가진 행렬로 설정
2. P Q.T값을 곱해 예측행렬 R을 계산하고 실제행렬 R에 해당되는 오류값을 계산한다. 
3. 이 오류값을 최소화할 수 있도록 P, Q를 적절한 값으로 각각 업데이트한다.
4. 만족할만한 오류값을 가질때까지 2,3을 반복하며 P,Q를 업데이트 한다.

앞에서와 유사하게 l2규제를 반영해 실제 r행렬값과 예측 r행렬값의 차이를 최소화하는 방향성을 갖고 p행렬과 q행렬에 업데이트 값을 반복적으로 수행하면서 최적화된 r행렬을 구하는 방식이 sgd 기반의 행렬분해이다.

분해하려는 원본행렬을 p,q로 분해한 뒤 다시 p,q.T의 내적으로 예측행렬을 만들고 원본행렬을 미정인 널값을 포함해 생성하고 분해행렬 p, q는 정규분포를 가진 랜덤값으로 초기화한다.




```python
import numpy as np
# 원본 행렬 R생성, 분해행렬 P,Q 초기화, 잠재요인차원 k는 3
R = np.array([[4, np.NaN, np.NaN, 2, np.NaN],
              [np.NaN, 5, np.NaN, 3,1],
              [np.NaN, np.NaN, 3,4,4],
              [5,2,1,2, np.NaN]])
num_users, num_items=R.shape
K=3
#P,Q행렬 크기 지정 및 정규분포를 가진 임의이 값으로 입력
np.random.seed(1)
P=np.random.normal(scale1./K, size=(num_users, K))
P=np.random.normal(scale1./K, size=(num_items, K))
```

다음으로 실제행렬과 예측행렬의 오차를 구하는 get rmse 함수를 생성하고자 한다. 이는 행렬값의 위치인덱스를 추ㅜㄹ해 이 인덱스에 있는 실제 R행렬과 분해된 p,q를 이용해 다시 조합된 예측행렬값의 rmse를 반환한다.



```python
from sklearn.metrics import mean_squared_error

def get_rmse(R,P,Q, non_zeros):
  error=0
  # 두개의 분해된 행렬 P, Q.T의 내적으로 예측행렬 생성
  full_pred_matrix = np.dot(P, Q.T)

  # 실제 R 행렬에서 널이 아닌 값의 위치인덱스 추출해 실제 R 행렬과 예측행렬의 RMSE 추출
  x_non_zero_ind [non_zero[0] for non_zero in non_zeros]
  y_non_zero_ind [non_zero[1] for non_zero in non_zeros]
  R_non_zeros = R[x_non_zero_ind, y_non_zero_ind]
  full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind]
  mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
  rmse = np.sqrt(mse)

  return rmse
```

```python
# sgd 기반의 행렬분해 수행
# r에서 널값을 제외한 데이터의 행렬인덱스를 추출하고 steps리는 업데이트 횟수 동안 업데이트 하면서 오류값을 출력
# R>0 인 행위치 열위치 값을 non_zeros에 저장
non_zeros [ (i,j,R[i,j]) for i in range(num_users) for j in range(num_items) if range(num_items) if R[i,j]>0]

steps = 1000
learning_rate = 0.01
r_lambda = 0.01

# update p,q matrix with sgd
for step in range(steps):
  for i,j,r in non_zeros:
    # 실제값과 예측값 차이인 오류값 구함
    eij = r-np.dpt(P[i,:], Q[j,:].T)
    # regularization을 ㅂ반영한 SGD 공식 적용
    P[i,:] = P[i,:] + learning_rate*(eij*Q[j,:] - r_lambda*P[i,:] )
    Q[j,:] = Q[j,:]+ learning_rate*(eij *P[i,:] -r_lambda*Q[j,:])
    rmse = get_rmse(R, P, Q, non_zeros)
    if (step % 50)==0:
      print('### iteration step: ', step, "rmse: ", rmse)
```
